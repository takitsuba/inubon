# 帰無仮説優位性検定とは
- ロジック
  - 「コインが公平である」と仮定する
    - $\theta=0.5$
    - コインを投げれば、だいたい半分の回数で表が出ることを期待
  - 実際に実験したところ、表の数がコインを投げた数の半分よりかなり大きい、または小さければ、「コインは公平である」と言う仮説(=帰無仮説)は棄却すべき。
  - では、この推論を正確にするためには？
    - 「起こりうる全ての結果」について、正確な確率を計算する
    - この確率を元に、「実験結果と同程度またはそれ以上に極端な結果が起きる確率」(=p値)を求める。
    - このp値がとても小さければ帰無仮説を棄却する。

- 「起こりうる全ての結果」
  - 上記の推論では
    - 帰無仮説のもとで生じる結果全体の空間を定義し、
    - 各結果の確率を計算する必要がある。
  - では「起こりうる全ての結果」は何で決まるか？
    - どのようにデータを集めようとしたか、によって決まる
  - データの集め方(サンプリング)の例
    - コインをN回投げようとした場合
    - Z回目の表が出るまでコインを投げようとした場合
    - 決められた時間内にコインを投げようとした場合
  - データの集め方によって、「起こりうる全ての結果」が変わる。


- 「実験結果と同程度またはそれ以上に極端な結果が起きる確率」(=p値)
  - 前述の通り、「起こりうる全ての結果」はデータの集め方によって変わる
  - そのためそれを元に計算するp値も変化する。
  - p値をより明確に定義すると、「**意図されたサンプリング（や検定の手続き）を用いる時に**、実験結果と同程度またはそれ以上に極端な結果が起きる確率」

- 想定する実験

# Nが固定されていた場合のp値の求め方
- N回中z回表が出る確率は下記
$$p ( z | N , \theta ) = \left( \begin{array} { c } { N } \\ { z } \end{array} \right) \theta ^ { z } ( 1 - \theta ) ^ { N - z }$$
- $\left( \begin{array} { c } { N } \\ { z } \end{array} \right)$は、${}_N\mathrm{C}_z$
- N=24の場合、zは0~24の25通りがありうる。それら全てについて、帰無仮説が正しい($\theta=0.5$)とした場合の確率を求め、グラフにしたのが図11.3の右側。

- 図11.3
<img src="./img/fig11_3.png">

- N=24の時、z=0となる確率、z=1となる確率、…、z=7となる確率を上記の式より計算したし合わせると、$p=0.032$となる(p値)
  - 帰無仮説のもとでぴったり$z/N$を得る確率ではなく、帰無仮説から期待される$z/N$ または、それよりも極端な結果を得る確率を求める。
  - 理由は、$z/N$という結果が期待するものからかなり外れているために帰無仮説を棄却するのだとすれば、もっと極端な値を持つ他の結果も、帰無仮説を棄却させるだろう。それゆえ、期待するものに対する
    - 実際の結果
    - または、それよりも極端な結果
  - を得る確率を知りたい。この全確率が「p値」と呼ばれる(片側p値)
  - 両側p値を求めるためには、片側p値は2倍される
  - p値が臨界値(慣習的に5%)よりも小さければ、帰無仮説を棄却する。(実際に起きた結果またはそれよりも極端なことが起きる確率がとても低いはずなのに、それが起きたということは、仮説が間違えている)
  

- 注意点
  - 帰無仮説が正しくても、5％の確率で帰無仮説を棄却するということ。
- 今回の例
  - 片側確率は$p=0.032 > 0.025$のため、**帰無仮説$\theta=0.5$を棄却しない**。
    - 帰無仮説を採択する、訳ではない。判断を保留。仮説は正しいかもしれないし、間違っているかもしれない。

# zが固定されていた場合のp値の求め方
- この場合、z回の表が出るのにN回のコイン投げが必要である確率はどのようになるか？
- N回目のコイン投げがz回目の表になることはわかっている。なぜなら、そのタイミングでコイン投げをやめるから。
- そのため、$N-1$回目のコイン投げでは$z-1$回の表を持つことになる。
- この時、求める確率は下記の通り

$$ \left.\begin{aligned} p ( N | z , \theta ) & = \left( \begin{array} { c } { N - 1 } \\ { z - 1 } \end{array} \right) \theta ^ { z - 1 } ( 1 - \theta ) ^ { N - z } \cdot \theta \\ & = \left( \begin{array} { c } { N - 1 } \\ { z - 1 } \end{array} \right) \theta ^ { z } ( 1 - \theta ) ^ { N - z } \\ & = \frac { z } { N } \left( \begin{array} { l } { N } \\ { z } \end{array} \right) \theta ^ { z } ( 1 - \theta ) ^ { N - z } \end{aligned} \right. $$

- 先ほどと同様に、全ての場合について確率を求め、グラフにする
  - $z=7$が固定されているため、$N$は7以上の全ての整数の可能性がある。
  - その結果を図示したものが図11.4の右側
- $z=7$で、$N=24$の時の確率、$N=25$の時の確率、...と足し合わせて行った時の確率が$p=0.017$となる。(p値)
- $p=0.017 < 0.025$となり、**帰無仮説を棄却する**。

# 帰無仮説を棄却しない→棄却するへと変わった
- $N=24$回コインを投げ、$z=7$回表が出たというデータは変わってない
- サンプリングを止める意図 (N固定、z固定)が異なると、p値も異なる、ということ。
- なぜ起きたか
  - 「表(または裏)の数が閾値をこえたらデータの収集を止める」のはデータにバイアスを与える。
    - ランダムに極端なデータが偶然得られた際に、データの収集をやめさせてしまい、その後のデータが得られない(はじめに表が7回出てしまうとそこで終了。その後のコイントスの結果は得られない。)
- 閾値$z$で止めることと閾値$N$で止めることと異なるp値を意味する。多くの実務家は極端な値が得られた時にデータの収集を止める。

# 時間を固定されていた場合
- アシスタントが「2分経過したからコイン投げをやめた」とする。
- この場合、閾値の時間に到達したからデータの収集をやめたのであり、閾値$N$や閾値$z$は存在しない。つまり$N$も$z$も決められていない
- このような場合、zとNの様々な組み合わせがどのように生じるかを考える。
- 正しい方法はないが、1つのやり方は「サンプルサイズ$N$を確率変数であるとみなすこと」
  - 2分間でできるコイントスの回数$N$は、大きいことも小さいこともある
- $N$の分布は例えばポアソン分布などとする。
  - ポアソン分布は$0$から$+\infty$までの整数をとる確率分布
  - パラメータは$\lambda$。非負の実数。これは分布の平均となる。
  - 今回の例であれば、$\lambda=24$とすれば$N$の値は通常24付近、それよりも大きかったり小さかったりすることもある
- これを表にしたものが図11.2の右側。

# 事前知識がある際の、NHSTとベイジアン分析の違い
- 想定する実験

  - 表：尖った方が上を向いて着地する場合
  - 裏：先の尖った方が地面についた場合

- 釘は裏が出やすいように偏っている、という強い事前の信念を持っている

# NHSTの場合
- NHSTの分析ではコインを投げるか、釘を投げるかは気にしない。
- 今回の場合であれば、図11.3のパターンと同様、帰無仮説を棄却しない。
- だが、釘については裏が出やすいという強い事前の信念がある。24回投げ、表が7回出た。それでも表裏が50：50であるという帰無仮説を棄却できないのは良いだろうか？

# ベイジアン分析の場合
- ベイズ統計では事前知識を表現することから分析を始める
  - 今回であれば、事前の経験から、頭の平たい釘は裏が出やすいと知っているため、その知識を事前分布で用いる
  - 事前分布は、科学的に納得できるものでなければならない
- 具体的にどうなるか
  - 釘について、同意を得た事前分布 $beta(\theta|2,20)$の場合(図11.7の左側)
    - 事後分布は$beta(\theta|2+7,20+17)$となる。
  - コインのように物体が公平であるという事前知識がある場合(図11.7の右側)
    - 架空の過去のサンプルを用いて事前の信念を表現する。$beta(\theta|10,10)$
    - さらに、架空のサンプルの以前に事前分布のプロトタイプが$beta(\theta|1,1)$であったとする
    - すると$beta(\theta|11,11)$の事前分布になる
    - 事後分布は$beta(\theta|11+7,11+17)$
  - 事後分布の95%HDIに、公平である(表が出る確率50%)ことが含まれているか含まれていないか、違いが出る

# 信頼区間と最高密度区間
- p値の危険を認識し、頻度主義の信頼区間(CI)であれば安心という意見もある
- しかし、「推定の不確かさを理解させる」という目標は、頻度主義の信頼区間では達成されない。
- ベイジアン分析ではその目標は達成される

# 信頼区間は意図に依存する
- NHSTの主な目的は、パラメータについての特定の「帰無」値が棄却されうるかを決めること
- 棄却できないパラメータ値の範囲が信頼区間
  - 95%信頼区間は、5%のフォールスアラームを許容する(両側の)有意性検定で棄却されない全ての$\theta$から構成される
- この信頼区間もp値と同様に、実験者の停止と検定の意図に依存する
  - 信頼区間はp値によって定義される
- 各信頼区間の算出は省略

# 信頼区間は分布ではない
- 信頼区間は単に2つの端点にすぎない
- よくある誤解は「$\theta$の値に関する確率分布を示している」というもの。そんなことはない
- 「信頼区間の中央にある$\theta$の値の方が、信頼区間の端の値よりも信頼できるはず」と考えてはいけない。
- 信頼区間に何らかの形の分布をつける方法はいくつかあるが、よくない。
  - ベイズはより直観を表現することができる。最高密度区間である

# ベイズ最高密度区間(HDI)
- 最高密度区間は信頼区間に似た、ベイズ推論の概念。
- NHSTの信頼区間と比べた時の、最高密度区間のメリット
  1. $\theta$の値の確信度から直接解釈できる。最高密度区間は$p(\theta|D)$に関するもので、私たちが知りたいこと
  2. 最高密度区間は実験者のサンプリングや検定の意図には依存しない
  3. 最高密度区間が事前の信念に応答する。どの程度事前の信念を変えるべきかが示されている

# 多重比較
- 9章の野球選手の例のように、様々な比較を行うことを考える。
- 複数の条件を比較するとき、NHSTの目標は「全体のフォールスアラーム率を望んだ最大値（例えば5％）よりも下に留めること」
  - これが達成されているかどうかは、比較の数や、実験者の意図の影響を受ける
- ベイジアン分析では条件を記述するパラメータに関して、事後分布は1つだけ
  - 比較の数や、実験者の意図の影響を受けない
  - よくわからん

# 実験単位のエラーに対するNHSTの修正(よくわからない)
- 複数の群(守備位置など)を比較することを考える
- 複数の比較を行おうとする意図はp値を増加させる。なぜなら、起こりうる結果の集合が拡大するから。
- 複数の検定を行う際には全体としてのフォールスアラーム率を求めること。
  - 全体のフォールスアラーム率を5％以下に抑えるためには、個々の比較に対して許容されるフォールスアラーム率を減らしたりする。(5％を比較回数で割るなど)
- ここで重要なのは、修正をどう行うかではなく、NHSTでは分析者が意図した比較の数やタイプによって、何らかの修正を行わなければならないということ。
- 例えば二人の分析者がいたとする。
  - 好奇心の強い分析者は多くの比較を行う。すると厳しい閾値を用いることになり、有意な結果を得づらくなる
  - 好奇心の弱い分析者は少ない比較を行う。優しい基準で得する。

# ベイズの事後分布は1つだけ
- ベイジアン分析では、停止や検定についての実験者の意図はデータの解釈に影響を与えない。
- 分析者は興味のあることを、事後分布をもとに調べることができる。
- そして事後分布は、私たちがどのようにそれを見ようとも、変わらない。意図の影響を受けない。
- また、NHSTと異なり、ベイズの事後分布は差の大きさの確信度を直接的に伝える。NHSTはその差が極端かどうかしかわからない。

# ベイジアン分析はどのようにフォールスアラームを減らすか
- フォールスアラームを完全に避けることはできない。無作為抽出されたデータは時々外れ値が偶然に得られるため。
- ベイジアンは観測されたデータと事前の知識があれば、私たちが行えるもっとも良い推論は事後分布であると考える。
- その上でベイジアン分析では、フォールスアラームの問題に、下記のように取り組む(わからん)
  - 事前知識をモデルの構造に取り入れる
  - 具体的には、異なる群が包括的な共通性を持つと分かっている場合、たとえそれらの具体的な処遇が異なるとしても、共通性を表現する包括的分布から抽出されたものとして、異なる群のパラメータを記述する。

# サンプリング分布が役に立つこと
- 観察されたデータセットからの推測において、サンプリング分布(仮想的な可能性の集合)は事後分布ほど有用ではない
  - サンプリング分布は、特定の仮説が成り立つとした場合に、生じうるデータの確率(コインの表が出る確率0.5とした場合に、データセットが生じる確率)
  - 特定のデータが得られた状況での、パラメータの確率は教えてくれない。
- そんなサンプリング分布でも、役立つことがある。それを見ていく。

# 実験を計画する際に役立つ
- 何か研究を行う際には、まず、データを得る前に実験を計画する
  - 行いたい実験があり、それから信頼できる結果を得るには、どの程度の観測値が必要かを考える
- 具体例
  - たとえば、選挙候補者Aの支持について、正確な事後信念を得るべく、母集団の調査を計画する
  - 支持率は$\theta \approx 0.6$と想定している。
  - この時、調査人数は $N=500$で良いか、考えたい。どのように行うか。
  - $\theta \approx 0.6$の乱数と、$N=500$で何度もシミュレーションを行い、データセットをえる。
  - すると、シミュレーションごとに、ベイズ事後分布を得る。
  - その事後分布からHDIの幅のような正確さを求める
  - こうすることで「HDIの幅」のサンプリング分布を得る
  - このサンプリング分布から、正確性という観点で$N=500$が適切な調査人数かを考えることができる。Nを増やしたり減らしたりする。

# モデルの予測を検討する(事後予測チェック)
- ベイジアン分析は、考慮している様々なパラメータやモデルの相対的な確信度を示すだけ
- 事後分布は、どのパラメータ値が相対的に悪くないかがわかるが、もっとも悪いパラメータ値が、実際にどの程度役に立つのかはわからない
- 具体例
  - イカサマコイン。99%表か裏が出る
  - コインを40回投げて、表が30回出たとする・
  - 99%表モデルの方が、99%裏モデルより事後確率が大きい。
  - 一方で、40回中30回しか表が出ないなら、99%表モデルはひどいモデルとも言える
  - この結果をどう評価するか
  

- ここで使えるのが、事後予測チェック
  - 事後分布の代表的なパラメータ値(上記であれば、表99%)をもとに、シミュレーションデータを作成する
  - このシミュレーションデータは、実際のデータに類似しているはず。
  - 類似していないなら、モデルを考え直す
    - 上記の例なら、40回中30回表が出たというデータは、表99％のシミュレーションデータと類似しないだろう
    - 表99%,裏99%のどちらかという考えが間違えていたと判断する
- 事後予測チェックによって、モデルが成功したか失敗したかの直観が得られ、どのように新しいモデルを導出すればいいかを考えることができる。
